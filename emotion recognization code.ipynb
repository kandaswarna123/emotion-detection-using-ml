```
import librosa
import numpy as np

def extract_features(audio_path, sample_rate=22050, n_mfcc=40):
    audio, sr = librosa.load(audio_path, sr=sample_rate)
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc)
    mfccs = np.mean(mfccs.T, axis=0)
    return mfccs
```

```
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import pandas as pd

# Assume features and labels are extracted
# * The features and labels list were empty before, populate them with your data. *
features = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]  # Example features - Replace with your actual data
labels = ['happy', 'sad', 'angry']  # Example labels - Replace with your actual data

# Convert to DataFrame for simplicity
data = pd.DataFrame(features)
data['label'] = labels

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:, :-1], data['label'], test_size=0.2, random_state=42)

# Train a model
clf = RandomForestClassifier()
clf.fit(X_train, y_train)

# Evaluate
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
